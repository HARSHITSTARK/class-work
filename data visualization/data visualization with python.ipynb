```python
pie_chart = pd.read_csv('C:\\Users\\harshit_kyal\\Documents\\Python Scripts\\8-1 Data Visualization\\8-1_Pie.csv')
```

```python
pie_chart
```

```python
pie_chart['Total Sales'] = pie_chart['Total Sales'].str.replace('$', '')#Replacing $ with blank
pie_chart['Total Sales'] = pie_chart['Total Sales'].str.replace(',', '')#Replacing , with blank
pie_chart['Total Sales'] = pie_chart['Total Sales'].astype(float)#Converting datatype to float
```

```python
explode = (0.1, 0, 0, 0)
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
plt.pie(pie_chart["Total Sales"], explode=explode, labels=pie_chart["Product"], colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

plt.show()
```
